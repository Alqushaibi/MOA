```python
# Plot surface of the objective function
def plot_surface(fobj, lb, ub, title):
    if lb is None or ub is None:
      ("Invalid function number. Please choose a valid function number.")
      return
    x = np.linspace(lb[0], ub[0], 100)
    y = np.linspace(lb[1], ub[1], 100)
    X, Y = np.meshgrid(x, y)
    Z = np.zeros_like(X)
    for i in range(X.shape[0]):
        for j in range(X.shape[1]):
            Z[i, j] = fobj([X[i, j], Y[i, j]])

    fig = plt.figure(figsize=(10, 6))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(X, Y, Z, cmap='viridis')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Objective Value')
    ax.set_title(title)
    plt.show()
```

```python

# Define Get_Functions_details function
def Get_Functions_details(F):
    switch_dict = {
        1: (-100, 100, 10, F1),
        2: (-10, 10, 10, F2),
        3: (-100, 100, 10, F3),
        4: (-100, 100, 10, F4),
        5: (-30, 30, 10, F5),
        6: (-100, 100, 10, F6),
        7: (-1.28, 1.28, 10, F7),
        8: (-500, 500, 10, F8),
        9: (-5.12, 5.12, 10, F9),
        10: (-32, 32, 10, F10),
        11: (-600, 600, 10, F11),
        12: (-50, 50, 10, F12),
        13: (-50, 50, 10, F13),
        14: (-65.536, 65.536, 2, F14),
        15: (-5, 5, 4, F15),
        16: (-5, 5, 2, F16),
        17: ([-5, 0], [10, 15], 2, F17),
        18: (-2, 2, 2, F18),
        19: (0, 1, 3, F19),
        20: (0, 1, 6, F20),
        21: (0, 10, 4, F21),
        22: (0, 10, 4, F22),
        23: (0, 10, 4, F23),
        # Add details for other benchmark functions here
    }
    if F not in switch_dict:
        print("Invalid function number.")
        return None, None, None, None

    lb, ub, dim, fobj = switch_dict.get(F, (None, None, None, None))
    if lb is not None and ub is not None:  # Check if lb and ub are provided
        if not isinstance(lb, (list, tuple)):
            lb = [lb] * dim
        if not isinstance(ub, (list, tuple)):
            ub = [ub] * dim
    return lb, ub, dim, fobj
```

```python

class MosquitoOptimizationAlgorithm:
    def __init__(self, num_mosquitoes, num_dimensions, max_iterations, search_range, F):
        self.num_mosquitoes = num_mosquitoes
        self.num_dimensions = num_dimensions
        self.max_iterations = max_iterations
        self.search_range = search_range
        self.best_solution = None
        self.best_fitness = np.inf
        self.convergence_curve = np.zeros(max_iterations)
        self.history = []
        self.F = F
        _, _, _, self.objective_function = Get_Functions_details(F)  # Properly unpack the tuple
        print(f'MOA is optimizing function {F}')

    def initialize_population(self):
        return np.random.uniform(-self.search_range, self.search_range,
                                 size=(self.num_mosquitoes, self.num_dimensions))

    def penetration(self, position, iteration):
        factor = 1 - (iteration / self.max_iterations)  # Decrease over time to increase exploitation
        return position * (0.9 * factor)

    def feeding(self, position, iteration):
        factor = 0.8 + (0.2 * (iteration / self.max_iterations))  # Increase over time to improve exploitation
        return position * factor

    def saliva_injection(self, position):
        return position * 0.95  # Suppress host's immune response and reduce inflammation

    def local_search(self, position):
        perturbation = np.random.uniform(-1, 1, self.num_dimensions) * 0.1
        new_position = position + perturbation
        if self.objective_function(new_position) < self.objective_function(position):
            return new_position
        return position

    def crossover(self, position1, position2):
        crossover_point = np.random.randint(0, self.num_dimensions)
        new_position = np.copy(position1)
        new_position[crossover_point:] = position2[crossover_point:]
        return new_position

    def optimize(self):
        positions_history = []

        population = self.initialize_population()

        for iteration in range(self.max_iterations):
            positions_history.append(population.copy())  # store copy of current positions
            for i in range(self.num_mosquitoes):
                population[i] = self.penetration(population[i], iteration)
                population[i] = self.feeding(population[i], iteration)
                population[i] = self.saliva_injection(population[i])
                population[i] = self.local_search(population[i])  # Apply local search

                fitness = self.objective_function(population[i])
                if fitness < self.best_fitness:
                    self.best_solution = population[i].copy()
                    self.best_fitness = fitness

                if iteration % 10 == 0 and i < self.num_mosquitoes - 1:
                    population[i] = self.crossover(population[i], population[i + 1])  # Apply crossover every 10 iterations

            self.convergence_curve[iteration] = self.best_fitness
            self.history.append(self.best_fitness)

            if iteration % 50 == 0:
              print(f'Iteration {iteration}: Best score = {self.best_fitness}')
            #   # Save scores to text file
            #   file_path = os.path.join(iterations_folder, f'MOA_iteration_F{F}.txt')
            #   with open(file_path, 'a+') as file:
            #       file.write(f'Iteration {iteration}: Best score = {self.best_fitness}\n')

        self.mean_score = np.mean(self.convergence_curve)
        self.worst_score = np.max(self.convergence_curve)

        return self.best_solution, self.best_fitness, self.convergence_curve, self.mean_score, self.worst_score, self.history, positions_history

# Example usage of the class
num_mosquitoes = 30
num_dimensions = 6
max_iterations = 1000
search_range = 10.0
F = 20  # Specify the function identifier


optimizer = MosquitoOptimizationAlgorithm(num_mosquitoes, num_dimensions, max_iterations, search_range, F)
best_solution_MOA, best_fitness_MOA, convergence_curve_MOA, mean_score_MOA, worst_score_MOA, history_MOA = optimizer.optimize()

print("Best solution:", best_solution_MOA)
print("Best fitness:", best_fitness_MOA)
print("Mean score:", mean_score_MOA)
print("Worst score:", worst_score_MOA)
print("History:", history_MOA)

# Plot convergence curve
plt.plot(convergence_curve_MOA)
plt.xlabel('Iteration')
plt.ylabel('Best score')
plt.title('Convergence Curve')
plt.grid(True)
plt.show()
```
